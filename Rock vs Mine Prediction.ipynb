```python
# import dependencies
import numpy as np # for arrays, as np is just a abbreviated
import pandas as pd # for processing steps
from sklearn.model_selection import train_test_split # split our data to training and test so no need for mannual
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score # find accuracy of our model

# loading the data set to a pandas dataframe
sonar_data = pd.read_csv('/content/sonar data.csv', header=None) # always mention header

# first 5 rows
sonar_data.head()

# num of rows and columns
sonar_data.shape

# show basic statistics (min, max, mean, etc)
sonar_data.describe()

# counts number of unique data, leading up to split
sonar_data[60].value_counts()

# finding mean for each value (m & r)
sonar_data.groupby(60).mean()

x = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]

'''
print(x)
print(y)
'''

# split data into training and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.1, stratify=y, random_state=1)

'''
print(x.shape, x_train.shape, x_test.shape)
output: (208, 60) (187, 60) (21, 60)
'''

# model training with logistic regression model
model = LogisticRegression()

# training the logistic regression model with training data
model.fit(x_train, y_train)

# EVAL MODEL
# accuracy on training data
training_data_accuracy = accuracy_score(model.predict(x_train), y_train)
print("Accuracy on training data: ", training_data_accuracy)

# accuracy on test data
test_data_accuracy = accuracy_score(model.predict(x_test), y_test)
print("Accuracy on test data: ", test_data_accuracy)

# making predicative system
input_data = (0.0131,0.0068,0.0308,0.0311,0.0085,0.0767,0.0771,0.0640,0.0726,0.0901,0.0750,0.0844,0.1226,0.1619,0.2317,0.2934,0.3526,0.3657,0.3221,0.3093,0.4084,0.4285,0.4663,0.5956,0.6948,0.8386,0.8875,0.6404,0.3308,0.3425,0.4920,0.4592,0.3034,0.4366,0.5175,0.5122,0.4746,0.4902,0.4603,0.4460,0.4196,0.2873,0.2296,0.0949,0.0095,0.0527,0.0383,0.0107,0.0108,0.0077,0.0109,0.0062,0.0028,0.0040,0.0075,0.0039,0.0053,0.0013,0.0052,0.0023)

# changing input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the np array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)

if prediction[0] == 'R':
  print("The object is a rock")
else:
  print("The object is a mine")





```
